```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

% matplotlib inline
% load_ext autoreload
% autoreload 2
```

### stage 1

```python
video_path = "../video/"
video_name = "1_0_Lounge_Iphone7_20033.MOV"
cap = cv2.VideoCapture(video_path + video_name)
video_len = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
fps = cap.get(cv2.CAP_PROP_FPS)
print("frame rate per second: {}".format(fps))

total_step = 8
stage1 = np.zeros((512,512,total_step))
for step in range(total_step):
    frame_no = 4 * int(fps) + step * int(fps) * 15
    cap.set(1, frame_no)
    # read a frame
    ret, frame = cap.read()
    # color to gray
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    # resize image
    gray = cv2.resize(gray, (1024, 1024))
    # crop image
    gray = gray[256:768, 256:768]
    # denoise
    gray = cv2.fastNlMeansDenoising(gray, None, 65, 2, 21)
    
    stage1[:,:,step] = gray
    
cap.release()
```

```python
f, axarr = plt.subplots(2,4, figsize=(40,20))
for step in range(total_step):
    axarr[step//4][step%4].imshow(stage1[:,:,step], cmap="gray")
```

```python
f, axarr = plt.subplots(1,3, figsize=(30,10))
axarr[0].plot(stage1[256,:,0])
axarr[1].plot(stage1[256,:,1])
axarr[2].plot(stage1[256,:,2])
```

```python
# histogram
f, axarr = plt.subplots(1,3, figsize=(30,10))
axarr[0].hist(stage1[:,:,0].ravel(),256,[0,256])
axarr[1].hist(stage1[:,:,1].ravel(),256,[0,256])
axarr[2].hist(stage1[:,:,2].ravel(),256,[0,256])
```
